```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Import data

```python
our_data = pd.read_csv("insurance.csv")
```

Plot data

```python
x = our_data.iloc[:,:1].values
y = our_data.iloc[:,-1].values

plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Insurance Cost')
plt.show()
```

Create linear regression without line of best fit 

```python
insurance_model = LinearRegression()
insurance_model.fit(x,y)
```

```python
y_pred = insurance_model.predict(x)

plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Insurance Cost')
plt.show()
```

Predict an unknown value and add line of best fit:

```python
unk_x = [[40]] 

x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = insurance_model.predict(x_pred)

plt.scatter(x,y,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Insurance Cost')
plt.show()

print("Predicted insurance cost is:", insurance_model.predict(unk_x))
```
